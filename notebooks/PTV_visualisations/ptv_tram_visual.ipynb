```python
import fiona
import folium
import geopandas as gpd
import pandas as pd
```

```python
gdf_METRO_TRAM = gpd.read_file('../../data/raw/PTV DATA/ll_gda94/filegdb/whole_of_dataset/victoria/PUBLIC_TRANSPORT.gdb', 
layer='PTV_METRO_TRAM_STOP')

df_TRAM_stops = pd.DataFrame({'STOP_ID': gdf_METRO_TRAM['STOP_ID'], 'geometry': gdf_METRO_TRAM['geometry'], 
'type': pd.Series(["TRAM" for x in range(len(gdf_METRO_TRAM.index))])})

map_tram = folium.Map(location=[-37.84, 144.94], zoom_start=10)
```

```python
# Create a geometry list from the GeoDataFrame
geo_df_list = [[point.xy[1][0], point.xy[0][0]] for point in df_TRAM_stops.geometry]

# Iterate through list and add a marker for each volcano, color-coded by its type.
i = 0
for coordinates in geo_df_list:
    # Place the markers with the popup labels and data
    map_tram.add_child(folium.Marker(location = coordinates,
                            popup='tram',
                            icon=folium.Icon(color = "blue")))
    i = i + 1
```

```python
gdf_METRO_TRAM_ROUTE = gpd.read_file('PTV DATA/ll_gda94/filegdb/whole_of_dataset/victoria/PUBLIC_TRANSPORT.gdb', 
layer='PTV_METRO_TRAM_ROUTE')

geoJSON_METRO_TRAM_ROUTE = gdf_METRO_TRAM_ROUTE[['ROUTE_ID', 'geometry']].to_json()

map_tram.add_child(folium.Choropleth(
    geo_data=geoJSON_METRO_TRAM_ROUTE,
    name='choropleth',
    line_color='red'
))
```
